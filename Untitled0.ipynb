```python
!pip install python-docx --upgrade
from google.colab import files
from docx import Document

uploaded = files.upload()
resume_filename = list(uploaded.keys())[0]

def extract_resume_text(file_path):
    doc = Document(file_path)
    return "\n".join([para.text for para in doc.paragraphs])

resume_text = extract_resume_text(resume_filename)
print("✅ Resume uploaded and parsed!")
job_description = input("Paste the job description here:\n")
!pip install langchain langchain-community openai python-docx
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate

llm = ChatOpenAI(temperature=0.3)

prompt = PromptTemplate(
    input_variables=["resume", "job"],
    template="""
You are a professional AI career coach. Review the resume and job description below.

Resume:
{resume}

Job Description:
{job}

Please suggest clear improvements for the resume to better match this job.
"""
)

final_prompt = prompt.format(resume=resume_text, job=job_description)
response = llm.predict(final_prompt)


print("📌 Resume Suggestions:\n")
print(response)


```
